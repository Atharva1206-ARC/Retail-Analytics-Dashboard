```python
import pandas as pd

# 1. Load dataset
df = pd.read_excel(r"C:\Users\HARSH\OneDrive\Desktop\Retail_Internship_Project\data\raw\Online Retail.xlsx")

# 2. Drop rows with missing CustomerID or Description
df.dropna(subset=["CustomerID", "Description"], inplace=True)

# 3. Remove rows with Quantity <= 0 or UnitPrice <= 0
df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]

# 4. Convert CustomerID to string, InvoiceDate to datetime
df["CustomerID"] = df["CustomerID"].astype(str)
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# 5. Add TotalPrice column
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

# 6. Save to cleaned CSV
output_path = r"C:\Users\HARSH\OneDrive\Desktop\Retail_Internship_Project\data\cleaned\retail_cleaned.csv"
df.to_csv(output_path, index=False)

print("✅ Cleaning complete. Data saved to:", output_path)
print("Shape:", df.shape)
df.head()
```

```python
import pandas as pd

# Load cleaned data
df = pd.read_csv(r"C:\Users\HARSH\OneDrive\Desktop\Retail_Internship_Project\data\cleaned\retail_cleaned.csv", parse_dates=["InvoiceDate"])

# Reference date for recency (max date in dataset)
snapshot_date = df["InvoiceDate"].max()

# Group by customer
customer_df = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (snapshot_date - x.max()).days,
    "InvoiceNo": "nunique",
    "TotalPrice": "sum",
    "Quantity": "sum",
    "Description": "nunique"
}).rename(columns={
    "InvoiceDate": "Recency",
    "InvoiceNo": "Frequency",
    "TotalPrice": "Monetary",
    "Quantity": "TotalItems",
    "Description": "DistinctItems"
})

# Average basket size and value
customer_df["AvgBasketSize"] = customer_df["TotalItems"] / customer_df["Frequency"]
customer_df["AvgBasketValue"] = customer_df["Monetary"] / customer_df["Frequency"]

# Drop TotalItems (intermediate)
customer_df.drop(columns=["TotalItems"], inplace=True)

# Reset index
customer_df.reset_index(inplace=True)

# Save engineered dataset
output_path = r"C:\Users\HARSH\OneDrive\Desktop\Retail_Internship_Project\data\cleaned\customer_features.csv"
customer_df.to_csv(output_path, index=False)

print("✅ Feature engineering complete. Saved to:", output_path)
customer_df.head()
```
